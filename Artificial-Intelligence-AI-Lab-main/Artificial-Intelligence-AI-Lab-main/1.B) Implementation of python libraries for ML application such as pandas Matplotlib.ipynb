# Matplotlib Library

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.plot([1,2,3,4],[1,4,9,16])
plt.show()
```

```python
plt.plot([1,2,3,4],[1,4,9,16])
plt.title("First Plot")
plt.xlabel('X label')
plt.ylabel('Y label')
plt.show()
```

```python
plt.figure(figsize=(15,5)) #for width and hieght
plt.plot([1,2,3,4],[1,4,9,16])
plt.title("First Plot")
plt.xlabel('X label')
plt.ylabel('Y label')
plt.show()
```
